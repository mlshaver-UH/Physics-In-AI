## Module 2 — Foundations of algorithms & building blocks

An **algorithm** is a finite, ordered, unambiguous sequence of steps that turns inputs into a correct output — a *language-independent* object that exists on paper before it exists in Python.

*(The word comes from the 9th-century scholar **Al-Khwarizmi**, whose name also gave us 'algebra'.)*

**Stages of an algorithm:** Input  →  Processing (ordered logical & arithmetic steps)  →  Output.

````python
# STAGES in one tiny function: INPUT -> PROCESSING -> OUTPUT
def axial_stress_MPa(force_N, area_mm2):     # INPUT (with units)
    stress = force_N / area_mm2              # PROCESSING (N/mm^2 = MPa)
    return stress                            # OUTPUT

print(axial_stress_MPa(1000.0, 4.0), 'MPa')
````

**Four pillars of computational thinking:** Decomposition · Pattern recognition · Abstraction · Algorithmic thinking.

**The three building blocks** — *any* algorithm is built from just these. Let's run one of each.

````python
# BUILDING BLOCK 1 - SEQUENCE: steps in order
force_N = 1000.0
area_mm2 = 4.0
sigma_MPa = force_N / area_mm2       # step 1
reserve = 250.0 - sigma_MPa          # step 2 (margin to yield, MPa)
print(f'stress = {sigma_MPa:.1f} MPa, margin to yield = {reserve:.1f} MPa')
````

````python
# BUILDING BLOCK 2 - SELECTION: choose a path with if / elif / else
yield_MPa = 250.0
if sigma_MPa >= yield_MPa:
    status = 'YIELDED'
elif sigma_MPa >= 0.8 * yield_MPa:
    status = 'WARNING'
else:
    status = 'SAFE'
print('status:', status)
````

````python
# BUILDING BLOCK 3 - ITERATION: repeat over many members
stresses_MPa = [120.0, 245.0, 300.0, 200.0]
for i, s in enumerate(stresses_MPa):
    print(f'member {i}: {s:6.1f} MPa -> {"YIELDED" if s >= yield_MPa else "SAFE"}')
````

**What makes an algorithm *good*:** finiteness (it terminates), definiteness (each step unambiguous), well-defined I/O, correctness (right answer for every valid input, including edge cases), and efficiency.

**Flowchart symbols:** rounded box = start/end · rectangle = process · diamond = decision · parallelogram = I/O · arrow = flow.

### Worked design — the SAFE/YIELDED classifier (before any code)
**Step 1 — SPEC (contract)**

| Symbol | Meaning | Units | Domain |
|---|---|---|---|
| `stresses` | stress per member | MPa | real |
| `yield_strength` | material yield strength | MPa | `> 0` |
| **out** `statuses` | 'SAFE' / 'YIELDED' | — | one per member |

**Step 2 — SKETCH** (pseudocode, three blocks only) is written as comments in the next cell; then we *Solve*.

````python
# SKETCH (pseudocode as comments)  ->  SOLVE
#   INPUT stresses, yield_strength
#   statuses <- empty list
#   FOR each s in stresses: IF s >= yield -> 'YIELDED' ELSE 'SAFE'
#   RETURN statuses

def classify_members(stresses, yield_strength):
    """Classify each member as SAFE/YIELDED. Pre: yield_strength > 0; MPa."""
    if yield_strength <= 0:
        raise ValueError('yield_strength must be > 0')
    return ['YIELDED' if s >= yield_strength else 'SAFE' for s in stresses]

print(classify_members([120.0, 245.0, 300.0], 250.0))
````

````python
# STEP 4 - STRESS-TEST: judge the output like an engineer
assert classify_members([250.0], 250.0) == ['YIELDED']   # boundary
assert classify_members([], 250.0) == []                 # edge: empty
try:
    classify_members([100.0], 0.0)                       # precondition
    print('FAIL: bad input accepted')
except ValueError as e:
    print('OK, rejected bad input:', e)
print('All stress-tests passed.')
````

## 4. Engineering problem — *design the algorithm* before coding

### Problem
A machined steel part leaves a furnace at **90 °C** and cools in still air at **25 °C**. 
By **Newton's law of cooling**, its temperature obeys the ODE
$$\frac{dT}{dt} = -k\,(T - T_\infty), \qquad T(0)=T_0$$
with cooling constant **k = 0.07 min⁻¹**. 
**Question:** how does the part's temperature evolve over the first hour, and can we trust a numerical solution?

This is a genuine *algorithm-design* task — there is no single formula to type; we must **discretize time and march the solution forward**. We walk the full design discipline.

### Step 1 — SPEC (the input/output contract)

| Symbol | Meaning | Units | Domain |
|---|---|---|---|
| `T0` | initial temperature | °C | real |
| `T_inf` | ambient temperature | °C | real |
| `k` | cooling constant | 1/min | `> 0` |
| `dt` | time step | min | `> 0` |
| `t_end` | total time | min | `> 0` |
| **out** `times, temps` | temperature history | min, °C | two equal-length lists |

### Step 2 — Constraints & invariants (the test oracle)
- The temperature must move **monotonically toward** `T_inf` (it cannot overshoot the ambient for this model).
- **Exact solution exists:** $T(t) = T_\infty + (T_0 - T_\infty)\,e^{-k t}$ — our gold standard for validation.
- **Stability limit (explicit Euler):** the step is stable only if `dt < 2/k`. A design constraint, discovered *before* trusting results.

### Step 3 — Decompose
1. Build the time grid. 2. Set the initial state. 3. **Iterate**: update T each step. 4. Collect history. 5. Validate.

### Step 4 — Method selection
We use **explicit (forward) Euler**: `T_next = T + dt * dT/dt`. Simplest possible scheme, first-order accurate — we choose it for transparency and will *quantify* its error rather than assume it.

### Step 5 — SKETCH (pseudocode)
```
ASSERT k > 0, dt > 0
t, T = 0, T0 ; times=[t]; temps=[T]
WHILE t < t_end:
    T = T + dt * (-k * (T - T_inf))     # Euler update
    t = t + dt
    append t, T
RETURN times, temps
```

````python
# Step 6 - SOLVE: turn the sketch into Python (pure Python: lists + a loop)
import math

def euler_cooling(T0, T_inf, k, dt, t_end):
    """Forward-Euler solution of dT/dt = -k (T - T_inf). Pre: k>0, dt>0."""
    if k <= 0 or dt <= 0:
        raise ValueError('k and dt must be > 0')
    t, T = 0.0, float(T0)
    times, temps = [t], [T]
    while t < t_end - 1e-9:
        T = T + dt * (-k * (T - T_inf))     # the building block: ITERATION
        t = t + dt
        times.append(t); temps.append(T)
    return times, temps

def exact_cooling(t, T0, T_inf, k):
    return T_inf + (T0 - T_inf) * math.exp(-k * t)

times, temps = euler_cooling(90.0, 25.0, 0.07, dt=1.0, t_end=60.0)
print(f'steps taken      : {len(times)-1}')
print(f'T(0)             : {temps[0]:.2f} C')
print(f'T(60) Euler      : {temps[-1]:.2f} C')
print(f'T(60) exact      : {exact_cooling(60.0, 90.0, 25.0, 0.07):.2f} C')
````

````python
# Step 7a - STRESS-TEST: accuracy vs the exact solution at sample times
times, temps = euler_cooling(90.0, 25.0, 0.07, dt=0.25, t_end=60.0)
print(f'{"t (min)":>8} {"Euler":>10} {"exact":>10} {"err":>9}')
for tcheck in (0, 10, 30, 60):
    i = min(range(len(times)), key=lambda j: abs(times[j]-tcheck))
    ex = exact_cooling(times[i], 90.0, 25.0, 0.07)
    print(f'{times[i]:8.2f} {temps[i]:10.3f} {ex:10.3f} {temps[i]-ex:9.4f}')
````

````python
# Step 7b - STRESS-TEST: convergence. Euler is 1st order -> halving dt ~halves error.
def max_error(dt):
    ts, Ts = euler_cooling(90.0, 25.0, 0.07, dt, 60.0)
    return max(abs(T - exact_cooling(t, 90.0, 25.0, 0.07)) for t, T in zip(ts, Ts))

prev = None
print(f'{"dt (min)":>9} {"max error":>11} {"ratio":>7}')
for dt in (2.0, 1.0, 0.5, 0.25):
    e = max_error(dt)
    ratio = (prev / e) if prev else float('nan')
    print(f'{dt:9.2f} {e:11.4f} {ratio:7.2f}')
    prev = e
print('ratio ~2 confirms first-order convergence (error is O(dt)).')
````

````python
# Step 7c - STRESS-TEST: STABILITY. Explicit Euler is stable only if dt < 2/k.
k = 0.07
print(f'stability limit dt < 2/k = {2/k:.1f} min')
ts, Ts = euler_cooling(90.0, 25.0, k, dt=35.0, t_end=210.0)   # dt > 2/k !
print('dt=35 (UNSTABLE) first temps:', [round(T,1) for T in Ts[:6]])
print('-> oscillates and diverges: the WRONG dt gives physically impossible temps.')
print('Lesson: choosing dt is part of the DESIGN, validated before trusting output.')
````

````python
# Step 7d - STRESS-TEST: edge cases
ts, Ts = euler_cooling(25.0, 25.0, 0.07, 1.0, 10.0)   # already at ambient
print('starts at ambient -> stays constant:', all(abs(T-25.0) < 1e-9 for T in Ts))
try:
    euler_cooling(90.0, 25.0, -0.07, 1.0, 60.0)        # k must be > 0
except ValueError as e:
    print('guarded bad k:', e)
````

````python
# Optional - visualize (Matplotlib). Safely skipped if not installed.
try:
    import matplotlib
    matplotlib.use('Agg')                 # headless-safe; in Jupyter omit this line
    import matplotlib.pyplot as plt
    ts, Ts = euler_cooling(90.0, 25.0, 0.07, 0.5, 60.0)
    exact = [exact_cooling(t, 90.0, 25.0, 0.07) for t in ts]
    plt.figure(figsize=(6,4))
    plt.plot(ts, Ts, 'o-', ms=3, label='Euler (dt=0.5)')
    plt.plot(ts, exact, '-', label='exact')
    plt.axhline(25, ls='--', color='gray', label='ambient')
    plt.xlabel('time [min]'); plt.ylabel('temperature [C]')
    plt.title('Cooling of a hot part'); plt.legend(); plt.tight_layout()
    plt.savefig('cooling.png', dpi=110)
    print('plotted -> cooling.png  (in Jupyter the figure shows inline)')
except ImportError:
    print('Matplotlib not installed -> pip install matplotlib')
````
